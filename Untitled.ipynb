```r
library(tidyverse)
library(repr)
library(tidymodels)
set.seed(1234)
```

```r
url_players<-"https://drive.google.com/uc?export=download&id=1M8-duCFfUh3MYjEn0j2WWRXzVMzTG97x"
players <- read_csv(url_players)

url_sessions <-"https://drive.google.com/uc?export=download&id=1DNxV79eDG7KkIWn6hIwgWeAeUN074SCS"
sessions <-read_csv(url_sessions)
```

```r
players_tidy<- players|>
               select(gender, experience, Age, played_hours)|>
               mutate(gender=as_factor(gender))|>
               mutate(experience= as_factor(experience))|>
               filter(!is.na(played_hours))

#Mean of qauntitative data from players data set
mean_values<- players|>
              select(Age, played_hours) |>
              summarise(
                    mean_age = mean(Age, na.rm=TRUE),
                  mean_played_hours = mean(played_hours, na.rm=TRUE))

mean_values_table <- as_tibble(mean_values)
mean_values_table

#Visualization of Data 
players_plot1<-players_tidy|>
               ggplot(aes(x=gender, y=played_hours,fill=experience))+
               geom_bar(stat="identity")+
               labs(x="Gender", y="Total Played Hours", fill="Experience", title="Gender vs Total Hours Played")
players_plot1

players_plot2 <-players_tidy|>
               ggplot(aes(x=Age,y=played_hours, fill= experience))+
               geom_bar(stat="identity")+
               labs(x="Age (years)", y="Total Hours Played", fill="Experience", title="Age vs Total Hours Played")
players_plot2 
```
